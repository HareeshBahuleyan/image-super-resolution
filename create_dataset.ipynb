```python
import os
import numpy as np
from PIL import Image
from tqdm import tqdm
from matplotlib import pyplot as plt
% matplotlib inline
```

### Create the dataset
Create high resolution and low resolution versions of the image dataset.
The original dataset has one of the dimensions as 500 and other between 300 and 500, so HR is created by resizing to 288 x 288. Let the LR be 72 x 72

```python
img_dir = '/collection/hpallika/deeplearning/research_work/fastai/image-super-resolution/data/mirflickr/'
hr_dir = '/collection/hpallika/deeplearning/research_work/fastai/image-super-resolution/data/hr_images/'
lr_dir = '/collection/hpallika/deeplearning/research_work/fastai/image-super-resolution/data/lr_images/'

hr_dims = (288, 288) # 4X times better resolution
lr_dims = (72, 72)

fnames = os.listdir(img_dir)
```

#### Low-res images:

```python
for f in tqdm(fnames):
    try:
        im = Image.open(img_dir + f)
        im = im.resize(lr_dims, resample = Image.ANTIALIAS)
        im.save(fp = lr_dir + f)
    except:
        pass
```

```python
im = Image.open(img_dir + f)
im = im.resize(lr_dims, resample = Image.ANTIALIAS)
plt.imshow(im)
```

#### High-res images:

```python
for f in tqdm(fnames):
    try:
        im = Image.open(img_dir + f)
        im = im.resize(hr_dims, resample = Image.ANTIALIAS)
        im.save(fp = hr_dir + f)
    except:
        pass
```

```python
im = Image.open(img_dir + f)
im = im.resize(hr_dims, resample = Image.ANTIALIAS)
plt.imshow(im)
```
